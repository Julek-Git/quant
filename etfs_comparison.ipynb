```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = yf.download(tickers=["SPY", "QQQ", "DIA", "DAX", "VGK"], interval="1d", period="10y").dropna()
```

```python
df["AvgPctChange"] = df["Close"].pct_change().T.mean()
```

```python
plt.figure(figsize=(12, 9))
plt.plot(df["AvgPctChange"], "g", linewidth=0.5, label="Average Percent Change")
plt.legend()
plt.grid()
plt.show()
```
